```python
#Take n numbers from the user and store them in a list.Print the list.
n=int(input("Enter number of elements :"))
numbers = []
for i in range(n):
    num = int(input("Enter elements :"))
    numbers.append(num)
print("List :",numbers)
```

```python
#From a list of numbers:Print the maximum,Print the minimum,Print the sum
n=int(input("Enter number of elements :"))
numbers = []
for i in range(n):
    num = int(input("Enter elements :"))
    numbers.append(num)
max_num = numbers[0]
min_num = numbers[0]
sum_num = 0
for i in range(n):
    if numbers[i]>max_num:
        max_num = numbers[i]
    if numbers[i]<min_num:
        min_num = numbers[i]
    sum_num += numbers[i]
    
print("Minimum number :",min_num)
print("Maximum number :",max_num)
print("Sum of numbers :", sum_num)
```

```python
#Take a list from the user and remove duplicate elements.
n=int(input("Enter number of elements :"))
numbers = []
for i in range(n):
    num = int(input("Enter elements :"))
    numbers.append(num)
new_list=[]
for i in range(n):
    if numbers[i] not in new_list:
        new_list.append(numbers[i])
print("List without duplicates :",new_list)
```

```python
#Reverse a list without using built-in reverse().
n=int(input("Enter number of elements :"))
numbers = []
for i in range(n):
    num = int(input("Enter elements :"))
    numbers.append(num)
start = 0
end = len(numbers)-1

while start<end:
    numbers[start],numbers[end]=numbers[end],numbers[start]
    start += 1
    end -= 1
print("Reversed list :", numbers)


    
```

```python
#Take a string and count how many vowels it has.
s= str("Aryan Mahanty")
s.split()
vowel="aeiouAEIOU"
count = 0
for i in range(len(s)):
    if s[i] in vowel:
        count += 1

print("No.of vowels :", count)
```

```python
#Take a string and check whether it is a palindrome or not.
n = int(input("Enter a number :"))
temp = n
r = 0
while n!=0:
    digit = n%10
    r = r*10 + digit
    n = n//10
if r == temp:
    print("It is palindrome")
else :
    print("It is not  palindrome")
    
```
