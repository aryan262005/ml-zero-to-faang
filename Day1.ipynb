```python
#Take an integer as input and calculate its factorial using a while loop.
n=int(input("Enter a number :"))
factorial = 1
while n!=0:
    factorial = factorial * n
    n -= 1
print("Factorial:" , factorial)
```

```python
#Take a number as input and check whether it is a prime number.
n=int(input("Enter a number to check prime or not :"))
if n<=1:
        print("It is not prime number")
else:
    for i in range (2 , n):
        if n%i == 0:
            print("It is not prime number")
            break
    else :
        print("It is prime number")
```

```python
#Take a number as input and print the reverse of the number.
n = int(input("Enter a number :"))
rev = 0
while n>0:
    digit = n%10
    rev = rev*10 + digit
    n = n//10
print("Reversed Number :", rev)
```

```python
#Given a list:[10, None, 25, None, 40, None]
#Print:Total number of elements,Count of missing values(None),Count of valid numbers
list1 = [10, None, 25, None, 40, None]
print("Number of elements : ", len(list1))

count = 0
for i in list1 :
    if i == None :
        count = count+1
print("Number of missing value :" ,count)

count1 = 0
for i in list1 :
    if i != None :
        count1 = count1+1
print("Number of valid nos :" ,count1)
```
